```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/josephgitau/September-intake/main/delivery_time.csv')
df.head()
```

#Simple linear regression

```python
plt.plot(df["Delivery Time"])
plt.xlabel("Delivery Time")
plt.ylabel("Sorting Time")
plt.title("Delivery Time Trend")
```

#  Problem statement - predict the delivery time based on the sorting time

```python
X=df[["Sorting Time"]]
y=df["Delivery Time"]

from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr.fit(X,y)

y_pred=lr.predict(X)

print("Model score: ", lr.score(X,y))
```

```python
# extract the model coefficients and intercept
print("Model coefficients: ", lr.coef_)
print("Model intercept: ", lr.intercept_)
```

```python
df['Predicted_Delivery Time'] = y_pred
df.head()
```
